# Potential step

```python
def potential(x):
    if x>x12:
        return V2
    return V1

def f(x, k1, k2, v):
    if x<0:
        return 1.*np.exp(1j*k1*x) + v[0]*np.exp(-1j*k1*x)
    return v[1]*np.exp(1j*k2*x)

def fref(x, k1, v):
    return v[0]*np.exp(-1j*k1*x)

def solution(V1, V2, energy):

    k1 = np.sqrt(2.*mass*(energy - V1))
    k2 = np.sqrt(2.*mass*(energy - V2))

    M = np.array([
            [1, -1],
            [-1, -k2/k1]
        ])
    inh = np.array([-1,-1])
    # find v such that M.x = inh
    v = np.linalg.solve(M, inh)

    xs = np.arange(-4, 4, .01)
    fs = [f(xsi, k1, k2, v) for xsi in xs]

    xs2 = np.arange(-4, 0., .01)
    fs2 = [fref(xsi, k1, [v[0],0,0,0]) for xsi in xs2]
    
    pots = [potential(xsi) for xsi in xs]

    return (xs, fs, xs2, fs2, pots)

def solution_RT(V1, V2, energy):

    k1 = np.sqrt(2.*mass*(energy - V1))
    k2 = np.sqrt(2.*mass*(energy - V2))

    M = np.array([
            [1, -1],
            [-1, -k2/k1]
        ])
    inh = np.array([-1,-1])
    # find v such that M.x = inh
    v = np.linalg.solve(M, inh)
  
#     print(np.abs(v[0])**2, ((k1-k2)/(k1+k2))**2)
#     print(k2/k1*np.abs(v[1])**2, 4*k1*k2/(k1+k2)**2)
    
    return (np.abs(v[0])**2, k2/k1*np.abs(v[1])**2)
#     return (((k1-k2)/(k1+k2))**2, 4*k1*k2/(k1+k2)**2)
```

# Tunnelling

```python
def potential(x, L):
    if x>0 and x<L:
        return barrier
    return 0

def f(L, x, k1, k2, k3, v):
    if x<0:
        return 1.*np.exp(1j*k1*x) + v[0]*np.exp(-1j*k1*x)
    elif x<L:
        return v[1]*np.exp(k2*x) + v[2]*np.exp(-k2*x)
    else:
        return v[3]*np.exp(1j*k3*x)

def fref(x, k1, v):
    return v[0]*np.exp(-1j*k1*x)

def solution_RT(L, barrier, energy):

    k1 = np.sqrt(2.*mass*(energy - potential(-1., L)))
    k2 = np.sqrt(2.*mass*(barrier - energy))
    k3 = k1

    M = np.array([
        [1,-1,-1,0],
        [-1j,-k2/k1,k2/k1,0],
        [0,np.exp(k2*L),np.exp(-k2*L),-np.exp(1j*k3*L)],
        [0,np.exp(k2*L),-np.exp(-k2*L),-1j*k3/k2*np.exp(1j*k3*L)]
    ])
    inhom = np.array([-1,-1j,0,0])
    v = np.linalg.solve(M, inhom)

    return v

def solution(L, barrier, energy):

    k1 = np.sqrt(2.*mass*(energy - potential(-1., L)))
    k2 = np.sqrt(2.*mass*(barrier - energy))
    k3 = k1

    M = np.array([
        [1,-1,-1,0],
        [-1j,-k2/k1,k2/k1,0],
        [0,np.exp(k2*L),np.exp(-k2*L),-np.exp(1j*k3*L)],
        [0,np.exp(k2*L),-np.exp(-k2*L),-1j*k3/k2*np.exp(1j*k3*L)]
    ])
    inhom = np.array([-1,-1j,0,0])
    v = np.linalg.solve(M, inhom)

    xs = np.arange(-3, 6, .01)
    fs = [f(L, xsi, k1, k2, k3, v) for xsi in xs]

    xs2 = np.arange(-3, 0., .01)
    fs2 = [fref(xsi, k1, [v[0],0,0,0]) for xsi in xs2]
    
    pots = [potential(xsi, L) for xsi in xs]

    return (xs, fs, xs2, fs2, pots)
```

```python
def potential(x, L):
    if x>0 and x<L:
        return barrier
    return 0

def f(L, x, k1, k2, k3, v):
    if x<0:
        return 1.*np.exp(1j*k1*x) + v[0]*np.exp(-1j*k1*x)
    elif x<L:
        return v[1]*np.exp(1j*k2*x) + v[2]*np.exp(-1j*k2*x)
    else:
        return v[3]*np.exp(1j*k3*x)

def fref(x, k1, v):
    return v[0]*np.exp(-1j*k1*x)

def solution(L, barrier, energy):

    k1 = np.sqrt(2.*mass*(energy - potential(-1., L)))
    k2 = 1j*np.sqrt(-2.*mass*(energy - barrier))
    k3 = k1

    A = np.array([[1,-1,-1,0],[-1,-k2/k1,k2/k1,0],[0,np.exp(1j*k2*L),np.exp(-1j*k2*L),-np.exp(1j*k3*L)],[0,k2*np.exp(1j*k2*L),-k2*np.exp(-1j*k2*L),-k3*np.exp(1j*k3*L)]])
    B = np.array([-1,-1,0,0])
    v = np.linalg.solve(A, B)

    xs = np.arange(-4, L+4, .01)
    fs = [f(L, xsi, k1, k2, k3, v) for xsi in xs]

    xs2 = np.arange(-4, 0., .01)
    fs2 = [fref(xsi, k1, [v[0],0,0,0]) for xsi in xs2]
    
    pots = [potential(xsi, L) for xsi in xs]

    return (xs, fs, xs2, fs2, pots)
```
